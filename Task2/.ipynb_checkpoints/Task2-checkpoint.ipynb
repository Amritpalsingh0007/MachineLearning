```python
import pandas as pd

df = pd.read_csv("/home/amritpal-singh/Downloads/weatherAUS.csv")
df.head(100)
```

```python
df.info()
```

```python
def checkingUnique():
    for i in df.columns:
        print(i, " : ", df[i].unique())
        print(len(df[i].unique()))
checkingUnique()
```

```python

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date', ascending=True).reset_index(drop=True)
```

```python
column_name = 'Date'  # Replace with the column you want to check for duplicates
duplicates = df[column_name].duplicated()
print(f'Number of duplicate values in {column_name}: {duplicates.sum()}')

# Display rows with duplicate values
duplicate_rows = df[df[column_name].duplicated(keep=False)]
print('Rows with duplicate values in column:')
print(duplicate_rows)
```
